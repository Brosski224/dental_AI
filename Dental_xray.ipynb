```python
from ultralytics import YOLO

# Load YOLO model
model = YOLO('yolov8n.pt')

# Train with CLAHE-augmented images
model.train(
    data="/content/drive/MyDrive/Dental OPG XRAY Dataset-20250306T161920Z-001/Dental OPG XRAY Dataset/Dental OPG (Object Detection)/Augmented Dataset/data.yaml",
    epochs=50,                # Increase epochs for better learning
    imgsz=640,                # Image size
    batch=16,                 # Batch size
    patience=5,               # Early stopping if no improvement in 5 epochs
    optimizer='AdamW',        # Better weight decay and generalization
    dropout=0.5,              # Reduces overfitting
    weight_decay=0.0005,      # Regularization
    device="CUDA"             # Use GPU (CUDA 12.4 in Colab)
)
```

```python
metrics = model.val()
print(metrics)
```

```python
results = model.predict("/content/drive/MyDrive/Dental OPG XRAY Dataset-20250306T161920Z-001/Dental OPG XRAY Dataset/Dental OPG (Classification)/Caries/112.jpg")
# Access the first Results object in the list (assuming there is at least one)
results[0].show()
```
